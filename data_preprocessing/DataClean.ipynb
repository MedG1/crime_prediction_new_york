# 3. Data preprocessing

## 3.1 Missing values

```python
ax = df.isna().sum().sort_values().plot(kind = 'barh', figsize = (9, 10))
plt.title('Percentage of Missing Values Per Column')
for p in ax.patches:
    percentage ='{:,.0f}%'.format((p.get_width()/df.shape[0])*100)
    width, height = p.get_width(), p.get_height()
    x = p.get_x() + width + 0.02
    y = p.get_y() + height/2
    ax.annotate(percentage, (x, y))
```

```python
cols = ['TRANSIT_DISTRICT', 'HOUSING_PSA', 'SUSP_AGE_GROUP', 'SUSP_SEX']
df = df.drop(cols, axis=1)
```

```python
df.head()
```

For the others columns with missing values we will try ti fill it with other values

```python
df['VIC_RACE'] = df['VIC_RACE'].fillna('UNKNOWN')
df['VIC_SEX'] = df['VIC_SEX'].fillna('U')
df['BORO_NM'] = df['BORO_NM'].fillna('UNKNOWN')
```

```python
df['ADDR_PCT_CD'] = df['ADDR_PCT_CD'].fillna(df['ADDR_PCT_CD'].mode()[0])
```

```python
df.isna().sum().sort_values()
```

## 3.2 Redundancy

We will drop columns with redundant info

```python
cols = ['CMPLNT_NUM', 'CMPLNT_TO_DT', 'CMPLNT_TO_TM','RPT_DT', 'OFNS_DESC', 'PD_DESC', 'JURIS_DESC','Lat_Lon', 'CMPLNT_TO_DT', 'CMPLNT_TO_TM','X_COORD_CD', 'Y_COORD_CD','PATROL_BORO' ]
df = df.drop(cols, axis=1)
```

```python
df.head()
```

```python
df.shape
```

## 3.3 Useless data

```python
cols = ['LOC_OF_OCCUR_DESC','PD_CD', 'KY_CD','JURISDICTION_CODE','SUSP_RACE','PREM_TYP_DESC','CRM_ATPT_CPTD_CD']
df = df.drop(cols, axis=1)
```

## 3.4 Spliting date & time data

```python
df.isna().sum().sort_values()
```

#### Change date columns to YYYY-MM-DD

```python
df['CMPLNT_FR_DT']= pd.to_datetime(df['CMPLNT_FR_DT'],errors='coerce')
df['CMPLNT_FR_TM'] = pd.to_datetime(df['CMPLNT_FR_TM'],errors='coerce')
```

```python
df = df[df['CMPLNT_FR_DT'].notna()]
df = df[df['CMPLNT_FR_TM'].notna()]
```

```python
df.isna().sum().sort_values()
```

#### Split date to year, month and day

```python
df['year'] = df['CMPLNT_FR_DT'].dt.year
df['month'] = df['CMPLNT_FR_DT'].dt.month
df['day'] = df['CMPLNT_FR_DT'].dt.day
df['hour'] =df['CMPLNT_FR_TM'].dt.hour
```

#### Add week day column

```python
df['weekday'] = pd.to_datetime(df['CMPLNT_FR_DT']).dt.day_name()
```

```python
df.drop(columns='CMPLNT_FR_DT',inplace = True)
df.drop(columns='CMPLNT_FR_TM',inplace = True)
```

```python
df.head()
```

```python
df.shape
```

## 3.5 Deal with victim age

```python
valid_age = ['UNKNOWN', '25-44', '<18', '45-64', '65+', '18-24']
df.loc[~df['VIC_AGE_GROUP'].isin(valid_age),'VIC_AGE_GROUP'] = 'UNKNOWN'
```

## 3.6 Encode data

```python
df.loc[df['PARKS_NM'].notnull(),'PARKS_NM'] = 1
df.loc[df['PARKS_NM'].isna(),'PARKS_NM'] = 0

df.loc[df['HADEVELOPT'].notnull(),'HADEVELOPT'] = 1
df.loc[df['HADEVELOPT'].isna(),'HADEVELOPT'] = 0

df.loc[df['STATION_NAME'].notnull(),'STATION_NAME'] = 1
df.loc[df['STATION_NAME'].isna(),'STATION_NAME'] = 0


df = df.rename({'PARKS_NM':'IN_PARK','HADEVELOPT':'IN_PUBLIC_HOUSING','STATION_NAME':'IN_STATION'},axis = 1)
```

```python
df['LAW_CAT_CD'] = df['LAW_CAT_CD'].replace(['felony','misdemeanor','violation'],[2,1,0])
df = df.rename({'LAW_CAT_CD':'CRIME_CLASS'},axis = 1)
```

```python
df['CRIME_CLASS'].value_counts()
```

```python
df.CRIME_CLASS = df['CRIME_CLASS'].replace(['FELONY','MISDEMEANOR',"VIOLATION"],[2,1,0])
```

```python
df.BORO_NM = df['BORO_NM'].replace(['BROOKLYN', 'STATEN ISLAND', 'BRONX', 'QUEENS', 'MANHATTAN','UNKNOWN'],[5,4,3,2,1,0])
```

```python
df.VIC_RACE = df['VIC_RACE'].replace(['WHITE', 'WHITE HISPANIC', 'BLACK','ASIAN / PACIFIC ISLANDER', 'BLACK HISPANIC',
       'AMERICAN INDIAN/ALASKAN NATIVE', 'OTHER', 'UNKNOWN'],[7,6,5,4,3,2,1,0])
```

```python
df.VIC_SEX = df['VIC_SEX'].replace(['F', 'M', 'D', 'E', 'U'],[4,3,2,1,0])
```

```python
df.isna().sum().sort_values()
```

```python
df = df[df['Latitude'].notna()]
df = df[df['Longitude'].notna()]
```
